```python
import pandas as pd
```

```python

# get api - https://www.alphavantage.co/support/#api-key

# documentation - https://www.alphavantage.co/documentation/
```

```python
apikey = ''
url = f'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=INTC&apikey={apikey}&outputsize=full&datatype=csv'

data = pd.read_csv(url)
print(data.info())
data.head()
```

### Описание колонок

timestemp - дата

open - цена открытия

high - самая высокая цена за период

low - самая низкая цена за период

close - цена закрытия

volume - объем торгов за период
